```python
def place_find(s) :
    # 접속 url
    url = 'http://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(s)
    
    # headers 입력
    headers = {
        # 카카오지도 접속
        "Authorization" : "KakaoAK f29c165521127a3107339d13a8bcc64b"
    }
    # API 요청
    palces = requests.get(url, headers=headers).json()['documents']
    try :
        palces = palces[0]
        jejuGwanGwang1_name  = palces['place_name']
        위도 = palces['y']
        경도 = palces['x']
        
        result = [jejuGwanGwang1_name, 위도, 경도]
        
        return result
    except:
        pass
```

```python
jeju_place = []

for x in top500['place']:
  a = place_find(x)
  jeju_place.append(a)

jeju_place
```

```python
jeju_place = [x for x in jeju_place if x is not None]
```

```python
import folium
jejuMap = folium.Map(location=[33.36, 126.52], zoom_start = 10)
```

```python
from folium.plugins import MarkerCluster

marker_cluster = MarkerCluster().add_to(jejuMap)
```

```python
for i in range(len(jeju_place)):
  folium.Marker(location = [jeju_place[i][1], jeju_place[i][2]],
                popup = jeju_place[i][0],
                icon = folium.Icon(color = 'blue', icon = 'ok'),
                ).add_to(marker_cluster)

jejuMap
```
