#### 극단적인 값
- 191 페이지 참고
- IQR : Inter Quartile Range (1사분위 ~ 3사분위 사이의 거리)
- 극단치 : 수염의 끝 (1사분위 - 1.5IQR / 3사분위 + 1.5IQR)

```python
##### IQR
## pct75 - 3사분면 / pct25 - 1사분면
pct75 = exam['math'].quantile(.75)
pct75
```

```python
pct25 = exam['math'].quantile(.25)
pct25
```

```python
IQR = pct75 - pct25
IQR
```

```python
##### 수염의 끝점
### 윗수염 3사분면 + 1.5IQR
pct75 + 1.5*IQR
```

```python
### 아래수염 1사분면 - 1.5IQR
pct25 - 1.5*IQR
```

```python
import re

p = re.compile("[^가-힣]")
moon = re.sub(p, " ", moon)
moon
```

```python
import konlpy
```

```python
hannanum = konlpy.tag.Hannanum()
```

```python
hannanum.nouns
```

```python
nouns = hannanum.nouns(moon)
nouns
```

```python
df_nouns = pd.DataFrame(nouns)
df_nouns
```

```python
df_nouns['count'] = df_nouns.iloc[:, 0].str.len()
df_nouns.rename({0 : 'word'}, axis = 1, inplace = True)
df_nouns
```

```python
## count 칼럼의 값이 2이상만 남기기
df_nouns = df_nouns[df_nouns['count'] >= 2]
df_nouns
```

```python
### 단어 빈도 구하기
top20 = df_nouns.groupby('word', as_index = False).agg('count').\
sort_values('count', ascending = False)[:20]

top20
```
