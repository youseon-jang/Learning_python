````python
#### 하위 디렉토리까지
def search(dirname):
    try:    
        filenames = os.listdir(dirname)
        for filename in filenames:
            full_filename = os.path.join(dirname, filename)
            if os.path.isdir(full_filename):
                search(full_filename)
            else:
                ext = os.path.splitext(full_filename)[-1]
                if ext == '.py':
                    print(full_filename)
    except PermissionError:
        pass
    
### os.path.isdir는 폴더인지 아닌지 알려줌
````

## 정규표현식(Regular Expression)

### 메타문자 

#### []

```
[abc] --> a, b, c 중 한 개의 문자와 매치

a - 매치
before - 매치(중간에 b가 있어도 됨)
dude - 매치되지 않음

[a-c] = [abc]
[0-5] = [012345]
[a-zA-z] = 모든 알파벳
[0-9] = 모든 숫자
```

#### ^

```
^가 대괄호 안에 있을 때에만 not임
[^0-9] = 숫자가 아닌 문자만 매치됨
```

### 자주 쓰는 정규 표현식

- \d : digits = [0-9]
- \D : not digits [^0-9]
- \s : space =  [t\n\r\f\v]
- \S : not \s = [^t\n\r\f\v]
- \w : alphabet + numeric = [a-zA-Z0-9]
- \W : not alphanumeric = [^a-zA-Z0-9]

```
. vs [.]
a.b ==> a와 b 사이에 줄바꿈 문자를 제외한 어떤 문자가 들어가도 모두 매치
a[.]b ==> a와 b사이에 Dot(.) 문자가 있으면 매치


*
ca*t ==> 문자 바로 앞에 있는 a가 0번 이상 반복되면 매치
ex) ct, cat, caaaat ....
ca가 반복되려면 (ca)*t 로 묶어야함


+
ca+t ==> 문자 바로 앞에 있는 a가 1번 이상 반복되면 매치
ex) cat, caaat, caaaat ....
*과 다르게 ct는 매치되지 않음


?
ab?c ==> 1번 이하 반복(있어도 되고 없어도 된다)
ex) abc, ac


{}
{m,n}
ca{2, 5}t ==> a가 2~5번 반복되면 매치(5도 포함)
ca{2, 2}t도 가능함 ca{2, 2}t = ca{2}t

```

````python
string = ["", "1", "12", "123", "1234", "abc", "1a"]
for i in string:
    print(re.sub("[0-9]?", "A", i)) 
### 숫자가 1번 이상이면 바꾸기
### 숫자 앞에 빈칸
### 문자 앞뒤
### 숫자는 바뀜
````

````python
li = ["", "1", "12", "123", "1234", "abc", "1a"]
for string in li:
    print(re.sub("[0-9]", "A", string), end = ",  ") ###  숫자가 1번 이상이면 바꾸기
````

````python
li = ["", "1", "12", "123", "1234", "abc", "1a"]
for string in li:
    print(re.sub("[0-9]+", "A", string), end = ",  ") ### 1개의 숫자가 1번 이상이면 바꾸기
````

````python
li = ["", "1", "12", "123", "1234", "abc", "1a"]
for string in li:
    print(re.sub("[0-9]?", "A", string), end = ",  ") 
### 숫자가 1개 이하면 바꾸기
### 1앞에 빈칸이 바뀌고 숫자 1이 바뀜
````

````python
li = ["", "1", "12", "123", "1234", "abc", "1a"]
for string in li:
    print(re.sub("[0-9]*", "A", string), end = ",  ") 
### 처음의 A는 빈칸
### 두 번째는 숫자바꾼 것
````

````python
li = ["", "1", "12", "123", "1234", "abc", "1a"]
for string in li:
    print(re.sub("[0-9]{4}", "A", string), end = ",  ") 
### 처음의 A는 빈칸
````

````python
li = ["", "1", "12", "123", "1234", "abc", "1a"]
for string in li:
    print(re.sub("[^0-9]", "A", string), end = ",  ") 
````

````python
li = ["", "1", "12", "123", "1234", "abc", "1a"]
for string in li:
    print(re.sub("^[0-9]", "A", string), end = ",  ") 
### ^이 밖에 있으면 시작을 의미함
### 시작이 숫자면  A로 바꾸기
````

````python
li = ["", "1", "12", "123", "1234", "abc", "1a"]
for string in li:
    print(re.sub("[0-9]$", "A", string), end = ",  ")
### 숫자로 끝나면 A로 바꾸기
````

- a로 시작하고 o로 끝나는 사이에 2개의 문자가 있는 경우 : a..o, [a]..[o], a.{2}o, ^a.{2}o$
- 알파벳 대문자로 시작하고 숫자로 끝나는 경우 : ^[A-Z].+[0-9]$, ^[A-Z]\S[0-9]$
- 123을 찾아서 바꾸기 : 123 ([123]은 1 or 2 or 3이라서 안됨)
- 문자들 중간에 123을 찾아서 바꾸기 : .+123.+, \S+123\S+, \D+123\D+

````python
li = ["a12o", "abco", "aBo", "a123o", "kbs", "123kbs", "KBS", "KB3", "abcKBS123"]
for string in li:
    print(re.sub("^[a]..[o]$", "A", string), end = ",  ")
````

````python
li = ["a12o", "abco", "aBo", "a123o", "kbs", "123kbs", "KBS", "KB3", "abcKBS123"]
for string in li:
    print(re.sub("a..o", "A", string), end = ",  ")
````

````python
li = ["a12o", "abco", "aBo", "a123o", "kbs", "123kbs", "KBS", "KB3", "abcKBS123"]
for string in li:
    print(re.sub("^[A-Z]\S[0-9]$", "A", string), end = ",  ")
````

````python
li = ["a12o", "abco", "aBo", "a123o", "kbs", "123kbs", "KBS", "KB3", "abcKBS123", "123123"]
for string in li:
    print(re.sub("123", "A", string), end = ",  ")
````

````python
li = ["a12o", "abco", "aBo", "a123o", "kbs", "123kbs", "KBS", "KB3", "abcKBS123", "123123"]
for string in li:
    print(re.sub("^\w123\w$", "A", string), end = ",  ")
````
