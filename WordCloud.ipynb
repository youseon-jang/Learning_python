```python
moon = open("speech_moon.txt", "r", encoding = 'utf-8').read()
```

```python
import re

p = re.compile("[^가-힣]")
moon = re.sub(p, " ", moon)
moon
```

```python
import konlpy
```

```python
hannanum = konlpy.tag.Hannanum()
```

```python
hannanum.nouns
```

```python
nouns = hannanum.nouns(moon)
nouns
```

```python
import pandas as pd

df_nouns = pd.DataFrame(nouns)
df_nouns
```

```python
df_nouns['count'] = df_nouns.iloc[:, 0].str.len()
df_nouns.rename({0 : 'word'}, axis = 1, inplace = True)
df_nouns
```

```python
## count 칼럼의 값이 2이상만 남기기
df_nouns = df_nouns[df_nouns['count'] >= 2]
df_nouns
```

```python
### 단어 빈도 구하기
top50 = df_nouns.groupby('word', as_index = False).agg('count').\
sort_values('count', ascending = False)[:50]

top50
```
