### 조건에 맞는 데이터만 추출하기

```python
exam = pd.read_excel("excel_exam.xlsx")
exam
```

```python
### class가 1인 거 검색
### exam['class'] == 1 는 True / False로 결과가 나오므로 검색 가능

exam[exam['class'] == 1]
```

```python
### math >= 50을 찾아라
exam[exam['math'] >= 50]
```

```python
### math 50점 이상이면서 1반

exam[(exam['math'] >= 50) & (exam['class'] == 1)]
```

- 조건을 각각 변수로 만들어주는 게 안전

```python
cond1 = exam['math'] >= 50
cond2 = exam['class'] == 1
```

```python
exam[cond1 & cond2]
```

## 데이터 정제 - 빠진 데이터, 이상한 데이터 제거하기

### 결측치 정제하기

```python
### 결측치 만들기
exam.iloc[0,2] = np.nan
exam.iloc[1, 3] = np.nan
exam.iloc[2, 4] = np.nan

df = exam.copy()
```

```python
df
```

- 결측치 존재 확인

```python
pd.isna(df)
```

```python
pd.isna(df).sum()
```

- 결측치 처리하기

```python
### 1_결측치가 있는 행 제거하기
df.dropna()

## 저장되지 않으므로 inplace 해야함
## 그냥 저장하면 index가 0부터 시작하지 않는다는 문제점 있음
```

```python
df.dropna(subset = 'math')  ### math column에 대해서만 NA 사라짐
```

```python
### math column 평균
m = np.mean(df['math'])
m
```

```python
np.mean(df.math)
```

```python
df['math'].mean()
```

```python
df.apply(np.mean)
```

- r과 다르게 NA가 있어도 결과 나옴
- r은 아예 에러 뜸

- 결측치 대체하기

```python
### 평균값으로 결측치 대체하기
df['math'] = df['math'].fillna(m)
```

```python
df
```

```python
df['english'].fillna(method = 'ffill')  ### 바로 앞의 값으로 저장
```

```python
df['english'].fillna(method = 'bfill', inplace = True)  ### 바로 뒤의 값으로 저장
df
```

```python
smean = np.mean(df['science'])
df['science'] = df['science'].fillna(smean)
df
```

```python
s = df['science'].mean()
df['science'].fillna(s, inplace = True)
df
```
